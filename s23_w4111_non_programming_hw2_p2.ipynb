````python
%%sql
#creat a function to create characterID
USE s23_w4111_hw2_xf2244;

DROP FUNCTION IF EXISTS compute_charID;

SET @c := 0;

CREATE FUNCTION compute_charID(characterName VARCHAR(64))
RETURNS VARCHAR(64)
READS SQL DATA
BEGIN
    DECLARE name CHAR(64);
    DECLARE num INT;
    DECLARE result VARCHAR(64);

    SET num = (@c := @c + 1);
    SET name = SUBSTRING_INDEX(characterName, ' ', 1);
    SET result = CONCAT(name, num);

    RETURN result;
END;
````

````python
%%sql

drop trigger if exists compute_characterID;


create trigger compute_characterID
    before insert
    on s23_w4111_hw2_xf2244.charactersFixed
    for each row
begin
    set new.characterID = compute_charID(new.characterName);
end;
````

- You will need to use SQL to create and load other tables with information from ```name_basics_all``` to enable you to create a view ```name_basics_all_fixed_view``` that recreates the data in ```name_basics_all```. The tables you create should have atomic columns, primary keys and foreign keys, etc.

````python
%%sql
describe name_basics_all
````

````python
%%sql
drop table if exists name_basics_all_fixed;
CREATE TABLE name_basics_all_fixed (
 nconst VARCHAR(10),
 primaryName VARCHAR(100),
 birthYear INT,
 deathYear INT,
 knownForTitles VARCHAR(100)
);
````

````python
%%sql
INSERT INTO name_basics_all_fixed(
    nconst, 
    primaryName, 
    birthYear, 
    deathYear, 
    knownForTitles
)
SELECT nconst, primaryName, birthYear, deathYear, knownForTitles
FROM name_basics_all
WHERE nconst NOT IN (SELECT nconst FROM name_basics_all_fixed);
````

````python
%%sql

drop table if exists name_basics_all_fixed;

CREATE TABLE name_basics_all_fixed AS
    SELECT 
    nconst, primaryName, birthYear, deathYear, knownForTitles,
    substring_index(substring_index(primaryProfession,',',1),',',-1) as Profession_1,
    substring_index(substring_index(primaryProfession,',',2),',',-1) as Profession_2,
    substring_index(substring_index(primaryProfession,',',3),',',-1) as Profession_3
    from name_basics_all
````

````python
%%sql
update name_basics_all_fixed set Profession_3 = NULL 
where Profession_3 = Profession_2;
update name_basics_all_fixed set Profession_2 = NULL 
where Profession_2 = Profession_1;
````

````python
%%sql
USE s23_w4111_hw2_xf2244;

DROP TABLE IF EXISTS titles;

CREATE TABLE titles AS
WITH 
one AS (
    SELECT 
        nconst,
        SUBSTR(knownForTitles, 1, LOCATE(',', knownForTitles) - 1) AS Title_1,
        SUBSTR(knownForTitles, LOCATE(',', knownForTitles) + 1) AS Title_2
    FROM name_basics_all_fixed
),
two AS (
    SELECT 
        nconst,
        Title_1,
        SUBSTR(Title_2, 1, LOCATE(',', Title_2) - 1) AS Title_2,
        SUBSTR(Title_2, LOCATE(',', Title_2) + 1) AS Title_3
    FROM one
),
three AS (
    SELECT 
        nconst, 
        Title_1,
        Title_2,
        SUBSTR(Title_3, 1, LOCATE(',', Title_3) - 1) AS Title_3,
        SUBSTR(Title_3, LOCATE(',', Title_3) + 1) AS Title_4
    FROM two
)
SELECT * FROM three;
````

````python
%%sql
DROP VIEW IF EXISTS name_basics_all_fixed_view;
CREATE VIEW name_basics_all_fixed_view AS
SELECT f.nconst, CONCAT(n.firstName, ' ', n.lastName) AS primaryName, f.birthYear, f.deathYear, 
CONCAT_WS(', ', p.Professional_1, p.Professional_2, p.Professional_3) AS primaryProfession,
CONCAT_WS(', ', t.Title_1, t.Title_2, t.Title_3, t.Title_4) AS knownForTitles
FROM name_basics_all_fixed f
LEFT JOIN names n ON f.nconst = n.nconst
LEFT JOIN professionals p ON f.nconst = p.nconst
LEFT JOIN titles t ON f.nconst = t.nconst;
````

````python
%%sql
DROP VIEW IF EXISTS name_basics_all_fixed_view;
CREATE VIEW name_basics_all_fixed_view AS
SELECT f.nconst, CONCAT(n.firstName, ' ', n.lastName) AS primaryName, f.birthYear, f.deathYear, 
CONCAT(IFNULL(p.Professional_1, ''), IF(IFNULL(p.Professional_1, '') <> '', ', ', ''), 
 IFNULL(p.Professional_2, ''), IF(IFNULL(p.Professional_2, '') <> '', ', ', ''),
 IFNULL(p.Professional_3, '')) AS primaryProfession,
CONCAT(IFNULL(t.Title_1, ''), IF(IFNULL(t.Title_1, '') <> '', ', ', ''),
 IFNULL(t.Title_2, ''), IF(IFNULL(t.Title_2, '') <> '', ', ', ''),
 IFNULL(t.Title_3, ''), IF(IFNULL(t.Title_3, '') <> '', ', ', ''),
 IFNULL(t.Title_4, '')) AS knownForTitles
FROM name_basics_all_fixed f
LEFT JOIN names n ON f.nconst = n.nconst
LEFT JOIN professionals p ON f.nconst = p.nconst
LEFT JOIN titles t ON f.nconst = t.nconst;
````
